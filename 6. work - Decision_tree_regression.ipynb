# Decision Tree Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Decision Tree Regression model on the whole dataset

```
# prompt: import Decision tree class

from sklearn.tree import DecisionTreeRegressor
AI_model = DecisionTreeRegressor()
AI_model.fit(X,y)
```

## Predicting a new result

```
pred = AI_model.predict([[2]])
pred
```

## Visualising the Decision Tree Regression results (higher resolution)

```
plt.scatter(x=X, y=y, color='blue')
plt.plot(X, AI_model.predict(X), color='red')
plt.title('Decision Tree Regresion')
plt.xlabel('Postion')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Decision Tree Regression results (higher resolution)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X,y,color='red')
plt.plot(X_grid, AI_model.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
